```python
# 5 Trovare velocità media di upload/download per la città Turin.
import csv
data = []
with open('./speedtest_short.csv', 'r') as myfile:
    reader = csv.DictReader(myfile)
    for value in reader:
        data.append(value)
media_download = 0
media_upload = 0
totale = 0 
for colonna in data:
    if colonna['CLIENT_CITY'] == 'Turin':
        media_download += int(colonna['DOWNLOAD_KBPS'])
        media_upload += int(colonna['UPLOAD_KBPS'])
        totale += 1
media_upload = int(media_upload) / int(totale)
media_download = int(media_download) / int(totale)
print 'UPLOAD_KBPS MEDIA TORINO: ' + str(media_upload)
print 'DOWNLOAD_KBPS MEDIA TORINO: ' + str(media_download)
```
